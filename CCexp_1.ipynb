3.IMAGE HANDLING

```python
from google.colab import files
from IPython.display import Image, display
from PIL import Image, ImageFilter
# 1. Upload the image
print("Please upload an image:")
uploaded = files.upload()
# 2. Process and display the image
for fn in uploaded.keys():
# Open the image
  img = Image.open(fn)
# --- Perform Image Manipulations ---
# Convert to grayscale
grayscale_img = img.convert('L')
# Blur the image more
blurred_img = img.filter(ImageFilter.BLUR)
blurred_img = blurred_img.filter(ImageFilter.BLUR)
# Rotate the image
rotated_img = img.rotate(90)
# --- Display the Images ---
# Set a new width for the displayed images
display_width = 200
# Display the images resized
print("\n--- Image Results ---")
print("Original Image:")
w, h = img.size
display(img.resize((display_width, int(h * display_width / w))))
print("Grayscale Image:")
w, h = grayscale_img.size
display(grayscale_img.resize((display_width, int(h * display_width / w))))
print("Blurred Image:")
w, h = blurred_img.size
display(blurred_img.resize((display_width, int(h * display_width / w))))
print("Rotated Image:")
w, h = rotated_img.size
display(rotated_img.resize((display_width, int(h * display_width / w))))
```

4.VIDEO HANDLING

```python
import cv2
from google.colab import files
import os

# Upload the video file
print("Please upload a video file:")
uploaded = files.upload()
video_path = list(uploaded.keys())[0]

# Create a directory to save the frames
output_folder = 'frames'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Open the video file
cap = cv2.VideoCapture(video_path)
frame_count = 0

# Read and save each frame
while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Save the frame as an image
    frame_filename = os.path.join(output_folder, f"frame_{frame_count:04d}.jpg")
    cv2.imwrite(frame_filename, frame)
    frame_count += 1

# Release the video capture object
cap.release()
print(f"\nExtraction complete. {frame_count} frames were extracted and saved in the '{output_folder}' directory.")
```
